# Below I am importing all of the required data to make our plots later. I am also dropping the unnamed columns and resetting our index to a datetime dtype of our Date columns in each DF.

```python
df = pd.read_csv('./Data/Bokeh_WNV_present.csv')
centroids = pd.read_csv('./Data/Bokeh_centroids.csv')
df_spray = pd.read_csv('./Data/Bokeh_spray.csv')

df_spray.drop(columns='Unnamed: 0', inplace = True)

centroids.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace = True)

df_spray['Date'] = pd.to_datetime(df_spray['Date'])
df_spray.set_index('Date', inplace = True)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace = True)
```

```python
# Thsi is the function thst I used to convert Longitude and Latitude into numbers that 
# Bokeh can actually plot with, this can be used on any longitude and latitude columns
# but they have to be put into a Location column as a tuple first for it to work.
def merc(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)
```

# Below we are generating a figure but we have not applied any glyphs, or data, to it yet.

```python
from bokeh.plotting import figure, show, output_notebook, reset_output, output_file# importing our basic Bokeh 
from bokeh.tile_providers import CARTODBPOSITRON
output_notebook()#Shows the map in the notebook environment
p = figure(x_range=(-9780000, -9745000), y_range=(5130000, 5160000),
           x_axis_type="mercator", y_axis_type="mercator")#Creating our figure
# range bounds supplied in web mercator coordinates
p.add_tile(CARTODBPOSITRON)# adding my tile and using the X range and y range above to generate Chicago
show(p)#show the plot
```

```python
# The Tools are pulled in as a string, they correspond to the items 
# on the upper left side of the plot, play with them to see what they do!
output_notebook()
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"
p = figure(x_axis_type="mercator", y_axis_type="mercator")
p = figure(title="2013 Traps with WNV present", tools=TOOLS)
p.add_tile(CARTODBPOSITRON)
p.circle(x = df['2013']['coords_x'],# THIS IS A GLYPH
         y = df['2013']['coords_y'], 
         size = 10, legend = 'Traps WNV present', color = 'red')
p.square(x = df_spray['2013']['coords_x'],# THIS IS ANOTHER GLYPH
         y = df_spray['2013']['coords_y'], 
         legend = 'Spray area 2013', color = 'blue', alpha =.05)
show(p)
```

```python
from bokeh.models import ColumnDataSource, HoverTool
# ColumnDataSource makes it easy to interact with the data in a pandas Dataframe interactively
output_notebook()
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"
source = ColumnDataSource(data=dict(
                        x=list(df['coords_x']), 
                        y=list(df['coords_y']),
                        WNV=list(df['WnvPresent']),
                        Num = list(df['NumMosquitos']),
                        sizes=list(df['NumMosquitos']),
                        Trap=list(df['Trap'])))
# Below is our hover tool, the right side of the tooltip must match the name on the left to work
hover = HoverTool(tooltips=[
    ("Trap", "@Trap"),
    ("WNV","@WNV"),
    ('Num', '@Num')
    
])
# range bounds supplied in web mercator coordinates
p = figure(x_axis_type="mercator", 
           y_axis_type="mercator",
           title="Traps based on # of mosquitos present w/ WNV",
           tools=[hover, TOOLS])
p.add_tile(CARTODBPOSITRON)
p.circle(x='x',
         y='y', 
         source=source,
         size= 'sizes',
         legend = 'Magnitude of mosquitos w/ WNV',
         line_color="#FF0000", 
         fill_color="#FF0000",
         fill_alpha=0.05)
p.legend.location = "bottom_left"
show(p)
```

```python
output_notebook()
source = ColumnDataSource(data=dict(
                        x=list(df['2013']['coords_x']), 
                        y=list(df['2013']['coords_y']),
                        WNV=list(df['2013']['WnvPresent']),
                        Num = list(df['2013']['NumMosquitos']),
                        sizes=list(df['2013']['NumMosquitos']),
                        Trap=list(df['2013']['Trap'])))
hover = HoverTool(tooltips=[
    ("Trap", "@Trap"),
    ("WNV","@WNV"),
    ('Num', '@Num')
    
])
# range bounds supplied in web mercator coordinates
p = figure(x_axis_type="mercator", 
           y_axis_type="mercator",
           title="2013 Traps based on # of mosquitos present w/ WNV, spray overlay, centroids",
           tools=[hover, TOOLS])
p.add_tile(CARTODBPOSITRON)
p.circle(x='x',
         y='y', 
         source=source,
         size= 'sizes',
         legend = 'Magnitude of mosquitos w/ WNV',
         line_color="#FF0000", 
         fill_color="#FF0000",
         fill_alpha=0.05)
p.circle(x = df_spray['2013']['coords_x'],
         y = df_spray['2013']['coords_y'], color = 'blue',
         legend = 'Spray locations 2013',
         alpha =.05)
p.triangle(x = centroids['coords_x'],
           y = centroids['coords_y'],
           legend = 'Geo-Statistical centers',
           size = 10, color = 'black')
# You can easily change the location of your legend with various intuitive inputs
p.legend.location = "bottom_left"
show(p)
```
